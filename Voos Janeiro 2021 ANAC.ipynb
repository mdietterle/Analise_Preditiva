# Projeto pontualidade de voos nos aeródromos do Brasil

O projeto em questão, tem por objetivo analisar a pontualidade das decolagens dos voos de passageiros nos aeródromos do Brasil. 

Queremos analisar se os atrasos ocorrem em uma taxa equivalente em todos os aeródromos, ou se algum se destaca dos demais. 

Queremos também verificar se os atrasos ocorrem em um determinado dia da semana e/ou em determinado período do dia.

O dataset é disponibilizado pela ANAC (Agência Nacional de Aviação Civil), e contém dados de todos os voos que tem por origem ou destino um aeródromo brasileiro. Um aeródromo é um local onde há uma pista de pouso, porém nem todos os aeródromos são aeroportos. Os que não possuem estrutura mínima para receber um voo são aeródromos porém não são aeroportos. Neste caso, não entram no relatório da ANAC e servem apenas para pousos e decolagens de aeronaves pequenas/particulares.

O relatório é disponibilizado mês a mês, portanto para fazermos uma análise dos dados precisamos baixar cada mês e juntar em um dataset só.

## Ler os dados dos datasets e juntar em um só

```
import pandas as pd

url = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20211.csv'
url2 = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20212.csv'
url3 = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20213.csv'
url4 = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20214.csv'
url5 = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20215.csv'
url6 = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20216.csv'
url7 = 'https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/VRA_20217.csv'

df1 = pd.read_csv(url, sep=';')
df2 = pd.read_csv(url2, sep=';')
df3 = pd.read_csv(url3, sep=';')
df4 = pd.read_csv(url4, sep=';')
df5 = pd.read_csv(url5, sep=';')
df6 = pd.read_csv(url6, sep=';')
df7 = pd.read_csv(url7, sep=';')

df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
df
```

## Eliminar voos que não nos interessam

Existem voos que não nos interessam de imediato, que são os com situação cancelado e os com status não informado. Por isso, iremos remover do dataset.

```
dfVooNCancelado = df[df['Situação Voo'] != 'CANCELADO']
dfVooNInformado = dfVooNCancelado[dfVooNCancelado['Situação Voo'] != 'NÃO INFORMADO']

dfVooNInformado.isna().sum()
```

## As únicas informações que nos interessam para o projeto são voos de passageiros

Os tipos de voo podem ser:

N - Doméstica Mista: para operações de transporte aéreo de passageiros ou mistas, em que todos os aeródromos envolvidos estejam situados simultaneamente em território brasileiro;

C - Doméstica Cargueira: para operações de transporte aéreo exclusivo de carga e/ou mala postal em que todos os aeródromos envolvidos estejam situados simultaneamente em território brasileiro;

I - Internacional Mista: para operações de transporte aéreo de passageiros ou mistas, em que ao menos um dos aeródromos envolvidos esteja situado em território estrangeiro;

G - Internacional Cargueira: para operações de transporte aéreo exclusivo de carga e/ou rede postal em que ao menos um dos aeródromos envolvidos esteja situado em território estrangeiro;

O dataset contém uma descrição como X, que não é padrão. Assume-se que esta descrição seja um tipo de voo que não se encaixa nas anteriores, como transferência de aeronave para outro aeródromo, ou algo semelhante.

Vamos manter no dataset somente as linhas que nos interessam

```
dfVoosPassageiros1 = dfVooNInformado[dfVooNInformado['Código Tipo Linha'] != 'X']
dfVoosPassageiros2 = dfVoosPassageiros1[dfVoosPassageiros1['Código Tipo Linha'] != 'C']
dfVoosPassageiros = dfVoosPassageiros2[dfVoosPassageiros2['Código Tipo Linha'] != 'G']
dfVoosPassageiros
```

## Criar colunas indicado voo com partida atrasada e chegada antecipada

Para podermos anonimizar os dados (veja mais sobre o assunto mais abaixo), iremos criar uma coluna contendo informações se o voo está pontual, atrasado ou adiantado.

Esta coluna nos ajudará a verificar se existe um aeródromo que repetidamente tem voos atrasados, ou se um dia da semana tem mais voos atrasados do que outros.

Os horários de partida e chegada real são obtidos com base no instante em que o trem de pouso deixa de tocar o solo (no caso de partida), e o momento em que o mesmo toca o solo (no caso de chegada).

Se partida real maior que partida prevista o voo está com partida atrasada.

Se chegada prevista maior que chegada real está com chegada antecipada.

```
import numpy as np

partidaAtrasada = [dfVoosPassageiros['Partida Prevista'] < dfVoosPassageiros['Partida Real'],
                   dfVoosPassageiros['Partida Prevista'] > dfVoosPassageiros['Partida Real']]
chegadaAntecipada = [dfVoosPassageiros['Chegada Prevista'] < dfVoosPassageiros['Chegada Real'],
                     dfVoosPassageiros['Chegada Prevista'] > dfVoosPassageiros['Chegada Real']]

situacao = ['Atrasado', 'Adiantado']

dfVoosPassageiros['Partida'] = np.select(partidaAtrasada, situacao, default='Pontual')
dfVoosPassageiros['Chegada'] = np.select(chegadaAntecipada, situacao, default='Pontual')

dfVoosPassageiros
```

## Criar colunas do dia da semana do voo e período do dia

Criar coluna com período do dia e dia da semana

Período do dia:

- Matutino ou Manhã: 6:00 às 11:59
- Vespertino ou Tarde: 12:00 (Meio-dia) às 17:59
- Noite ou Noturno 18:00 às 23:59
- Madrugada ou Sembrol: 00:00 (Meia-noite) às 05:59

```
# Extraindo dia da semana da data de partida e criando uma coluna com a informação
dfVoosPassageiros['Dia da Semana'] = pd.to_datetime(dfVoosPassageiros['Partida Prevista']).dt.day_name()

dfVoosPassageiros
```

```
# Extrair a hora da partida prevista e encaixar em um dos períodos, criando uma coluna para armazenar a informação
def hr_func(ts):
    return ts.hour

dfVoosPassageiros['horas'] = pd.to_datetime(dfVoosPassageiros['Partida Prevista']).apply(hr_func)
b = [0,4,8,12,16,20,24]
l = ['Madrugada', 'Inicio Manhã','Manhã','Tarde','Tarde / Início Noite','Noite']
dfVoosPassageiros['Parte do dia'] = pd.cut(dfVoosPassageiros['horas'], bins=b, labels=l, include_lowest=True)
dfVoosPassageiros

# hours = pd.to_datetime(dfVoosPassageiros['Partida Prevista']).dt.hour
# hours
# pd.concat([dfVoosPassageiros, pd.DataFrame(hours, index=dfVoosPassageiros.index)], axis = 1)
# dfVoosPassageiros
```

## Anonimzar dados

Por conta da LGPD precisamos eliminar algumas colunas desnecessárias. Queremos que não seja possível identificar unicamente um voo, para impedir que alguma empresa aérea seja prejudicada.

O nosso estudo será baseado em dados do aeroporto em geral, não com algum outro tipo de dado.

O ICAO da empresa aérea é um identificador único da empresa aérea em termos mundiais, ou seja, podemos identificar a empresa de qualquer parte do mundo com base nele.

O número do voo também pode levar à identificação da empresa, pois com o cruzamento do mesmo com o aeródromo temos a identificação do voo.

Por fim, os horários de partida e chegada também fazem com que o seu cruzamento com o aeródromo identifica a empresa aérea, pois há somente uma partida ou chegada em um determinado horário (por questões óbvias de segurança somente um voo pode ocupar a pista por vez).

Por isso, iremos remover as colunas ICAO Empresa aérea, Código de Autorização (DI), horários de partida e chegada (tanto prevista como real).

```
dfVoosPassageiros.drop(['ICAO Empresa Aérea'], axis='columns', inplace=True)
dfVoosPassageiros.drop(['Código Autorização (DI)'], axis='columns', inplace=True)
dfVoosPassageiros.drop(['Partida Prevista'], axis='columns', inplace=True)
dfVoosPassageiros.drop(['Partida Real'], axis='columns', inplace=True)
dfVoosPassageiros.drop(['Chegada Prevista'], axis='columns', inplace=True)
dfVoosPassageiros.drop(['Chegada Real'], axis='columns', inplace=True)
dfVoosPassageiros.drop(['Código Justificativa'], axis='columns', inplace=True)


dfVoosPassageiros
```

```
# Juntar dados dos aeroportos com latitude/longitude pelo ICAO
aero = pd.read_csv('https://raw.githubusercontent.com/mdietterle/Analise_Preditiva/main/airport-codes_csv.csv', sep=',')
aero.rename(columns = {'ident' : 'ICAO Aeródromo Origem'}, inplace = True)
df_left = pd.merge(dfVoosPassageiros, aero, on='ICAO Aeródromo Origem', how='inner')
# aero
# df_left.dropna()
# df_left.isna().sum()
# df_left['coordinates']
df_left[['long','lat']] = df_left['coordinates'].str.split(',',expand=True)
dfSul = df_left.loc[(df_left['iso_region']=='BR-PR') | (df_left['iso_region']=='BR-SC') | (df_left['iso_region']=='BR-RS')]
dfSul
```
